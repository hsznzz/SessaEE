### **3.1.2 SEE using K-means clustering definition**

```python
def SEE(arg1):
    C09CA01 = df[df["ATC"] == arg1].copy()
    C09CA01 = C09CA01.sort_values(by=["pnr", "eksd"])
    
    # Creating previous prescription date
    C09CA01["prev_eksd"] = C09CA01.groupby("pnr")["eksd"].shift(1)
    C09CA01.dropna(subset=["prev_eksd"], inplace=True)
    
    # Select one random row per patient
    Drug_see_p1 = C09CA01.groupby("pnr").sample(1, random_state=1234)[["pnr", "eksd", "prev_eksd"]]
    
    # Calculate event interval
    Drug_see_p1["event_interval"] = (Drug_see_p1["eksd"] - Drug_see_p1["prev_eksd"]).dt.days
    
    # ECDF Calculation
    ecdf = np.sort(Drug_see_p1["event_interval"].values)
    y = np.arange(1, len(ecdf)+1) / len(ecdf)
    dfper = pd.DataFrame({"x": ecdf, "y": y})
    dfper = dfper[dfper["y"] <= 0.8]  # Retaining 80% of ECDF
    
    # ECDF Visualization
    fig, axs = plt.subplots(1, 2, figsize=(12, 6))
    axs[0].scatter(dfper["x"], dfper["y"], color='red')
    axs[0].set_title("80% ECDF")
    axs[0].set_xlabel("Event Interval (Days)")
    axs[0].set_ylabel("ECDF")
    
    axs[1].scatter(ecdf, y, color='blue')
    axs[1].set_title("100% ECDF")
    axs[1].set_xlabel("Event Interval (Days)")
    axs[1].set_ylabel("ECDF")
    plt.tight_layout()
    plt.show()
    
    ni = dfper["x"].max()
    Drug_see_p2 = Drug_see_p1[Drug_see_p1["event_interval"] <= ni]
    
    # Density Plot
    density = gaussian_kde(np.log(Drug_see_p2["event_interval"]))
    x_vals = np.linspace(min(np.log(Drug_see_p2["event_interval"])), max(np.log(Drug_see_p2["event_interval"])), 100)
    plt.figure(figsize=(8, 5))
    plt.plot(x_vals, density(x_vals), color='green')
    plt.title("Log(Event Interval) Density Plot")
    plt.xlabel("Log(Event Interval)")
    plt.ylabel("Density")
    plt.show()
    
    # Silhouette Score for Optimal Clustering
    scaled_data = StandardScaler().fit_transform(dfper[["x"]])
    best_k = 2
    best_score = -1
    for k in range(2, 10):
        km = KMeans(n_clusters=k, random_state=1234).fit(scaled_data)
        score = silhouette_score(scaled_data, km.labels_)
        if score > best_score:
            best_score = score
            best_k = k
    
    # K-means Clustering
    kmeans = KMeans(n_clusters=best_k, random_state=1234).fit(scaled_data)
    dfper["cluster"] = kmeans.labels_
    
    # Cluster Summary
    cluster_summary = dfper.groupby("cluster")["x"].agg(["min", "max", "median"]).reset_index()
    
    # Assign Cluster to Data
    Drug_see_p1 = Drug_see_p1.merge(cluster_summary, left_on="event_interval", right_on="min", how="left")
    Drug_see_p1["cluster"] = Drug_see_p1["cluster"].fillna(0).astype(int)
    
    return Drug_see_p1
```

### **3.1.2 SEE assumption** [REQUIRES FIX] 

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd


def see_assumption(arg1):
    if not isinstance(arg1, pd.DataFrame):
        print("❌ Error: Input is not a DataFrame.")
        return np.nan

    if arg1.empty:
        print("⚠ Warning: No valid data for boxplot.")
        return np.nan

    if "pnr" not in arg1.columns or "eksd" not in arg1.columns:
        print("❌ Error: Missing required columns ('pnr', 'eksd').")
        return np.nan

    # Ensure eksd is a datetime column
    arg1["eksd"] = pd.to_datetime(arg1["eksd"], errors="coerce")

    arg1 = arg1.sort_values(by=["pnr", "eksd"]).copy()
    arg1["prev_eksd"] = arg1.groupby("pnr")["eksd"].shift(1)
    arg1["p_number"] = arg1.groupby("pnr").cumcount() + 1

    # Check for single-prescription patients
    patient_counts = arg1["pnr"].value_counts()
    single_prescription_patients = patient_counts[patient_counts == 1].index

    if len(single_prescription_patients) == len(patient_counts):
        print("⚠ All patients have only one prescription. Boxplot is not meaningful.")
        return np.nan

    # Filter out first prescriptions since they have no previous duration
    Drug_see2 = arg1[arg1["p_number"] > 1].copy()

    print(f"🔍 Total Rows Before Filtering: {arg1.shape[0]}")
    print(f"🔍 Total Rows After Filtering (p_number > 1): {Drug_see2.shape[0]}")

    if Drug_see2.empty:
        print("⚠ Warning: No valid Duration values for boxplot.")
        return np.nan

    Drug_see2["Duration"] = (Drug_see2["eksd"] - Drug_see2["prev_eksd"]).dt.days
    Drug_see2["p_number"] = Drug_see2["p_number"].astype("category")

    # Plot boxplot
    plt.figure(figsize=(8, 6))
    sns.boxplot(x="p_number", y="Duration", data=Drug_see2)
    plt.title("Boxplot of Duration by p_number")
    plt.xlabel("Prescription Sequence")
    plt.ylabel("Duration (Days)")

    # Add median line
    medians_of_medians = Drug_see2.groupby("pnr")["Duration"].median().median()
    plt.axhline(y=medians_of_medians, color="red", linestyle="dashed", alpha=0.5)

    plt.tight_layout()
    plt.show()

    return plt
```

### **3.4 Alternative Clustering Implementation (DBSCAN)**

```python
from sklearn.cluster import DBSCAN

def perform_dbscan_clustering(data, eps=0.5, min_samples=5):
    """
    Applies DBSCAN clustering on the given dataset.

    Parameters:
        data (pd.DataFrame): The dataset to cluster.
        eps (float): The maximum distance between two samples for them to be considered as in the same neighborhood.
        min_samples (int): The number of samples in a neighborhood for a point to be considered a core point.

    Returns:
        pd.DataFrame: The dataset with an added 'Cluster' column.
    """
    scaler = StandardScaler()
    scaled_data = scaler.fit_transform(data)

    dbscan = DBSCAN(eps=eps, min_samples=min_samples)
    cluster_labels = dbscan.fit_predict(scaled_data)

    data["Cluster_DBSCAN"] = cluster_labels
    return data
```
